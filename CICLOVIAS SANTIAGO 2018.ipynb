```python
datosNombres = datosTabla.filter(items=['nombre'])
datosNombres
```

```python
def puntosLinea(listaCoordenadas):
    puntos = []    
    listaFinal = []
    puntoUnico = []
    
    for coordenadas in listaCoordenadas:
        coordenadas = coordenadas.split(",0")
        coordenadas = coordenadas[:-1]
        puntos.append(coordenadas)
        
    for lista in puntos:
        listaOrdenada = []
        for punto in lista:
            punto = punto.split(",")
            #print(punto[0],punto[1])
            puntoUnico = [float(punto[1]),float(punto[0])]
            listaOrdenada.append(puntoUnico)
        listaFinal.append(listaOrdenada)        
        
    return listaFinal 
```

```python
listaPuntos = puntosLinea(datosTabla.coordenadas)
listaPuntos
```

```python
datosNombres = datosNombres.to_dict(orient='split')
```

```python
listaCiclovias = pandas.DataFrame(
    {'nombres': datosNombres,
     'puntos': listaPuntos
    })
listaCiclovias
```

```python
listaCiclovias = listaCiclovias.to_dict(orient='split')
```

```python
listaCiclovias = listaCiclovias['data']
```

```python
listaCiclovias[0][1]
```

```python
import folium
import html

mapaCiclovias = folium.Map(
    location = [-33.4372, -70.6506],
    tiles='Stamen Terrain',
    zoom_start = 12
)

for nombres, puntos in listaCiclovias:
    nombres = html.escape(str(nombres))
    nombres = nombres.strip('[').strip(']').strip("&#x27;*").strip()
    folium.PolyLine(puntos, color = 'red', opacity=0.7, popup=nombres).add_to(mapaCiclovias)

mapaCiclovias
```
