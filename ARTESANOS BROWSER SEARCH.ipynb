```python
import re
listaArtesanos = []



for data in informacion:
    #print(data)
    #print("------------------")
    pattern1 = r"Comuna:\xa0[\w ]*"
    comuna = re.findall(pattern1, str(data))[0]
    print(comuna)
    comuna = comuna.replace(u'\xa0', u' ')
    print(comuna)
    comuna = comuna[8:]
    print(comuna)
    pattern2 = r"Región:\xa0[\w ]*"
    region = re.findall(pattern2, str(data))[0]
    region = region.replace(u'\xa0', u' ')
    region = region[8:]
    nombre = data.select('h2')[0].text
    datos = [nombre, region, comuna]
    listaArtesanos.append(datos)
    
listaArtesanos
```

```python
import pandas
datosArtesanos = pandas.DataFrame(listaArtesanos)
```

```python
datosArtesanos
```

```python
datosArtesanos = datosArtesanos.to_dict(orient='split')
```

```python
datosArtesanos = datosArtesanos['data']
```

```python
geocodes = {}

def ubicacionesFiltradas(comuna,region):   
    buscar = (comuna+', '+region+", CHILE")
    x = geocodes.get(buscar, '')
    if x == '':
        x = geocoder.google(buscar).latlng
        geocodes[buscar] = x
    if x == None:
        x = geocoder.komoot(buscar).latlng
        geocodes[buscar] = x
    if x == None:
        x = geocoder.arcgis(buscar).latlng
        geocodes[buscar] = x
    return x
```

```python
coordenadas = []
for nombre, region, comuna in datosArtesanos:
    lugar = ubicacionesFiltradas(comuna,region)
    coordenadas.append(lugar)
    
coordenadas
```

```python
datosTabla1 = pandas.DataFrame(coordenadas)
datosTabla2 = pandas.DataFrame(datosArtesanos)
```

```python
datosFinales = pandas.concat([datosTabla2,datosTabla1], axis=1)
datosFinales
```

```python
datosFinales.keys()
```

```python
tamanoCirculos = datosFinales[2].value_counts()
tamanoCirculos = tamanoCirculos.to_dict()
tamanoCirculos
```

```python
datosArtesanos = datosFinales.to_dict(orient='split')
datosArtesanos = datosArtesanos['data']
datosArtesanos
```

```python
datosArtesanos
```

```python
import geocoder
import folium

mapaArtesanos = folium.Map(
    location = [-33.4550553, -70.5411876],
    tiles='Stamen Terrain',
    zoom_start = 4
)


for nombre, region, comuna, latitud, longitud in datosArtesanos:
    folium.Circle(
        location=[latitud,longitud],
        popup = 'Comuna: '+comuna+', Artesanos: '+str(tamanoCirculos[comuna]),
        fill = True,
        fill_color = 'blue',
        fill_opacity = 0.1,
        color = 'blue',
        radius = tamanoCirculos[comuna]*1000
    ).add_to(mapaArtesanos)

mapaArtesanos.save('mapa.html')
mapaArtesanos
```

```python
mapaArtesanosIcono = folium.Map(
    location = [-33.4550553, -70.5411876],
    tiles='Stamen Terrain',
    zoom_start = 4
)

def limitar(num, minimum=5, maximum=20):
  return max(min(num, maximum), minimum)

icon_url = 'http://www3.gobiernodecanarias.org/medusa/wiki/images/thumb/8/83/Bernegal.png/300px-Bernegal.png'

for nombre, region, comuna, latitud, longitud in datosArtesanos:
    tamano = limitar(tamanoCirculos[comuna])
    icono = folium.features.CustomIcon(icon_url,icon_size=(tamano*28/5,tamano*30/5))
    folium.Marker(location=[latitud,longitud],
                  popup='Comuna: '+comuna+', Artesanos: '+str(tamanoCirculos[comuna]),
                  icon=icono
                 ).add_to(mapaArtesanosIcono)
               
mapaArtesanosIcono.save('mapaiconos.html')
mapaArtesanosIcono
```
